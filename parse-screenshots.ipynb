```python
# Importing packages

import pytesseract as tess
from PIL import Image
import pandas as pd
import numpy as np
```

```python
# Loading Image object
img = Image.open('cache/buy-orders/green_wood.png')
type(img)
```

```python
# Upscaling image to make parse more accurate
newsize = (3840, 2160)
img = img.resize(newsize)
```

```python
# Extracting text from image
# text = tess.image_to_string(img)
green_wood_df = tess.image_to_data(img, output_type=tess.Output.DATAFRAME)
# text = tess.image_to_alto_xml(img)
```

```python
for text in green_wood_df['text']:
    print(text)
```

```python
green_wood_df = green_wood_df.dropna()
```

```python
for text in green_wood_df['text']:
    print(text)
```

```python
price_list = []
price_sum = 0
price_dirty_average = 0
price_min = 9999999
price_max = 0

# goes row by row, and if it can turns data into float
for text in green_wood_df['text']:

    try:
        price = float(text)

        if (price > 0):

            price_list.append(price)
            price_sum += price

            if price > price_max:
                price_max = price

            if price < price_min:
                price_min = price

    except:
        pass
    
price_dirty_average = price_sum / float(len(price_list))
print(price_min, price_max)
```

```python
# Getting data to use to filter out non-price numbers
price_q1, price_q3 = np.percentile(price_list, 25), np.percentile(price_list, 75)   # Q1 and Q3
price_iqr   = price_q1 - price_q3       # IQR

print(price_q1, price_q3, price_iqr)
```
